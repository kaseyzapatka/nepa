```python
from datasets import load_dataset
import pandas as pd

# 1. Load dataset
ds = load_dataset("PNNL/NEPATEC2.0", data_files="EA/*/*.jsonl", split="train")


def get_val(d, key, default=""):
    """Safely extract .value fields with nested structure."""
    try:
        return d.get(key, {}).get("value", default)
    except:
        return default

# 2. Flatten records
records = []

for item in ds:
    project = item.get("project", {})
    process = item.get("process", {})
    documents = item.get("documents", [])

    # ---- PROJECT FIELDS ----
    project_id = project.get("project_ID", "")
    project_title = get_val(project, "project_title")
    project_sector = get_val(project, "project_sector")
    project_type = get_val(project, "project_type")
    project_description = get_val(project, "project_description")
    project_sponsor = get_val(project, "project_sponsor")
    project_location = get_val(project, "location")

    # ---- PROCESS FIELDS ----
    process_family = get_val(process, "process_family")
    process_type = get_val(process, "process_type")
    lead_agency = get_val(process, "lead_agency")

    # ---- DOCUMENTS ----
    for doc in documents:
        metadata = doc.get("metadata", {})

        doc_meta = metadata.get("document_metadata", {})
        file_meta = metadata.get("file_metadata", {})

        document_id = get_val(doc_meta, "document_ID")
        document_type = get_val(doc_meta, "document_type")
        document_title = get_val(doc_meta, "document_title")
        prepared_by = get_val(doc_meta, "prepared_by")
        ce_category = get_val(doc_meta, "ce_category")

        file_id = get_val(file_meta, "file_ID")
        file_name = get_val(file_meta, "file_name")
        section_title = get_val(file_meta, "section_or_volume_title")
        main_document = get_val(file_meta, "main_document")
        total_pages = get_val(file_meta, "total_pages")
        file_provider = get_val(file_meta, "file_provider")

        # ---- PAGES ----
        for page in doc.get("pages", []):
            page_number = page.get("page number", None)
            page_text = page.get("page text", "")

            records.append({
                # PROJECT
                "project_id": project_id,
                "project_title": project_title,
                "project_sector": project_sector,
                "project_type": project_type,
                "project_description": project_description,
                "project_sponsor": project_sponsor,
                "project_location": project_location,

                # PROCESS
                "process_family": process_family,
                "process_type": process_type,
                "lead_agency": lead_agency,

                # DOCUMENT
                "document_id": document_id,
                "document_type": document_type,
                "document_title": document_title,
                "prepared_by": prepared_by,
                "ce_category": ce_category,

                # FILE
                "file_id": file_id,
                "file_name": file_name,
                "section_title": section_title,
                "main_document": main_document,
                "total_pages": total_pages,
                "file_provider": file_provider,

                # PAGE
                "page_number": page_number,
                "page_text": page_text,
            })

# Convert to DataFrame
df_ea = pd.DataFrame(records)
df_ea.head()
```
